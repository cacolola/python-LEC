### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
sentinels = {'a1206': [' '], 'a1207': [' ']}
df=pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
#将CEPS.csv数据读入python
df
```

```python
a=df.sample(n=1000)#随机抽取1000个数据
a
```

```python
x1=a['a1206']
print(x1)
x2=a['a1207']
print(x2)#根据问卷从数据中挑选'a1201'和'a1202"这两个连续型变量
```

```python
x1.mean()
```

```python
x2.mean()
```

```python
count=0
mean_1206={}
mean_1207={}
while count<30:#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
    df_a=df.sample(n=1000)
    mean_1206_a=df_a['a1206'].mean()
    mean_1206[count+1]= mean_1206_a
    mean_1207_a=df_a['a1207'].mean()
    mean_1207[count+1]= mean_1207_a
    count=count+1
print(mean_1206)
print(mean_1207)
```

```python
ind=np.arange(len(mean_1206.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_1206.values(), width, 
                color='Y', label='a1206均值')
ax.set_ylabel('同意程度')
ax.set_title('随机抽取a1206题样本的同意程度均值')
ax.set_xticks(ind)
ax.set_xticklabels(mean_1206.keys())
ax.legend()
ax.set_ylim(1,5)

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')



plt.show()
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

+ 步骤如下：
+ 取 c25"你对自己未来有没有信心" 做因变量y
+ 取 c12“目前的成绩” 、b35"父母对你未来是否有信心"、b09“家庭经济条件”做自变量

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd

sentinels = {'c25': [' '], 'c12': [' '], 'b35': [' '], 'b09': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
sdf=df.sample(n=1000)
sdf
```

```python
data1= pd.DataFrame({
    'y': sdf.c25,
    'x1': sdf.c12,
    'x2': sdf.b35,
    'x3': sdf.b09
})
#构建自变量
data1=data1.dropna(axis=0,how='any')#删除缺失值
model_x = ['x1','x2','x3']
X = data1.loc[ :,model_x].values
#构建因变量
y = data1['y'].values
#加入截距项
X_model = sm.add_constant(X)
print(X)
print(y)
print(X_model)
model = sm.OLS(y, X) 
#拟合ols回归
results = model.fit()
results.params

#输出结果
print(results.summary())
```

+ 由回归结果可知，调整的判定系数为0.965，即多元回归模型的拟合程度较好
  + F检验统计量的值为8853，显著大于F检验的临界值，即模型的线性关系显著
  + 三个自变量的系数分别为 0.1102、0.6613、0.2460
  + 由P值可知，三个自变量的P值结果都小于0.05，即三个自变量的系数都是显著的
  + 从线性关系检验和回归系数显著性检验都能看出，模型的拟合程度较好，且各个自变量对因变量的影响都是显著的
+ 综上可知，对自己未来有没有信心，受到目前的成绩、父母对自己未来的信心和家庭经济条件的影响
