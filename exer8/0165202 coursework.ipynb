### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False 
data = pd.read_csv(r'CEPS.csv',engine='python')
data.head
```

```python
x=data.sample(n=1000)
x1=x['b15d1'].convert_objects(convert_numeric=True)
x2=x['b15f1'].convert_objects(convert_numeric=True)
#选取 b09：课外看书时间   b12：游戏时间
```

```python
x1.mean()
```

```python
x2.mean()
```

```python
count=0
mean_b15d1={}
mean_b15f1={}
while count<30:
    data_x=data.sample(n=1000)
    mean_b15d1_a=data_x['b15d1'].convert_objects(convert_numeric=True).mean()
    mean_b15d1[count+1]= mean_b15d1_a
    mean_b15f1_a=data_x['b15f1'].convert_objects(convert_numeric=True).mean()
    mean_b15f1[count+1]= mean_b15f1_a
    count=count+1
print(mean_b15d1)
print(mean_b15f1)
```

```python
ind=np.arange(len(mean_b15d1.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_b15d1.values(), width, 
                color='red', label='课外看书时间')
ax.set_ylabel('样本均值')
ax.set_title('日平均课外看书时间')
ax.set_xticks(ind)
ax.set_xticklabels(mean_b15d1.keys())
ax.legend()
ax.set_ylim(0,2)

def autolabel(rects, xpos='center'):
    xpos = xpos.lower() 
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
plt.show()
```

```python
ind=np.arange(len(mean_b15f1.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_b15f1.values(), width, 
                color='black', label='游戏时间')
ax.set_ylabel('样本均值')
ax.set_title('日平均游戏时间')
ax.set_xticks(ind)
ax.set_xticklabels(mean_b15f1.keys())
ax.legend()
ax.set_ylim(0,2)

def autolabel(rects, xpos='center'):
    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
plt.show()
```

```python
b15d1=[]
for i in mean_b15d1.values():
    b15d1.append(i)
b15f1=[]
for i in mean_b15f1.values():
    b15f1.append(i)
average={'课外看书时间':b15d1,'游戏时间':b15f1}
frame = pd.DataFrame(average,index=mean_b15d1.keys())

print('样本均值：')
print(frame.mean())
print('\n样本标准误：')
print(frame.std())
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
%matplotlib inline
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
#导入数据
sentinels = {'c12': [' '], 'b01': [' '], 'b06': [' '], 'b07': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
df=df.sample(n=1000)   #随机抽取1000个样本数据
df
```

```python
#建立数据框
data = pd.DataFrame({
    'x1': df.b01,
    'x2': df.b06,
    'x3': df.b07,
    'y': df.c12})
#父母的教育程度和家庭状况对孩子成绩的影响
data=data.dropna(axis=0,how='any') 
```

```python
model_x= ['x1', 'x2','x3']
X = data.loc[ :,model_x].values
y=data['y'].values
```

```python
model = sm.OLS(y, X)    #拟合普通最小二乘回归
results = model.fit()
```

```python
results.params  #各变量系数
```

```python
print(results.summary())
```

```python
1.模型拟合指标
y=“你目前的成绩”
x1=“是否独生子女”x2=“你妈妈的教育水平”，x3=“你爸爸的教育水平”
2.模型的显著性检验结果
F统计量的p值小于0.05，R方为0.85，模型是合理的，对因变量的的波动因素的解释度达85%
3.变量的系数
分别为：0.96508224,  0.19474391,  0.1526462 
4.各系数的显著性检验结果
x1,x2,x3的系数的p值均接近于0，小于0.05，拒绝原假设，说明自变量与因变量之间是显著性相关的。
5.对模型结果的解释
原数据是服从正态分布的
线性回归模型y=0.96508224*x1+0.19474391*x2+0.1526462*x3
可以看到，非独生子女成绩往往会更高，妈妈的教育水平比父亲的教育水平对孩子成绩的影响力更大，
毕竟母亲一般较父亲与孩子相处的时间更多，对孩子的影响也较大。
```
