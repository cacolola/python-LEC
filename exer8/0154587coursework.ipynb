### 样本均值分布的统计试验

+ 请将CEPS.csv数据读入python
+ 请从中随机抽取1000个数据
+ 请根据问卷从数据中挑选两个连续型变量（likert量表可以近似看作连续变量）
+ 计算这两个连续变量的均值
+ 重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
+ 绘制这30个样本均值的直方图
+ 计算均值的均值和标准误

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
#读取数据
data = pd.read_csv(r'CEPS.csv',engine='python')
data
```

```python
x=data.sample(n=1000)
x1=x['a1801'].convert_objects(convert_numeric=True)
x2=x['b2501'].convert_objects(convert_numeric=True)
x1.mean()
```

```python
x2.mean()
```

```python
count=0
mean_a1801={}
mean_b2501={}
while count<30:#重复随机抽取—计算均值这个过程30次，得到两个变量30个样本均值
    data_x=data.sample(n=1000)
    mean_a1801_a=data_x['a1801'].convert_objects(convert_numeric=True).mean()
    mean_a1801[count+1]= mean_a1801_a
    mean_b2501_a=data_x['b2501'].convert_objects(convert_numeric=True).mean()
    mean_b2501[count+1]= mean_b2501_a
    count=count+1
print(mean_a1801)
print(mean_b2501)
```

```python
ind=np.arange(len(mean_a1801.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_a1801.values(), width, 
                color='g')
ax.set_xticks(ind)
ax.set_xticklabels(mean_a1801.keys())
ax.legend()
ax.set_ylim(1,4)

def autolabel(rects, xpos='center'):
    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')

ind=np.arange(len(mean_b2501.values()))
width=0.5
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, mean_b2501.values(), width, 
                color='r')
ax.set_xticks(ind)
ax.set_xticklabels(mean_b2501.keys())
ax.legend()
ax.set_ylim(1,4)

def autolabel(rects, xpos='center'):
    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 1, 'right': 1, 'left': 1}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 2*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')
plt.show()
```

```python
#计算30个样本均值的均值和标准误
a1801=[]
for i in mean_a1801.values():
    a1801.append(i)
    
b2501=[]
for i in mean_b2501.values():
    b2501.append(i)
    
average={'七天内的沮丧程度':a1801,
         '与母亲关系':b2501,
        }
frame = pd.DataFrame(average,index=mean_a1801.keys())

print(frame.mean())
print(frame.std())
```

### 回归分析

+ 请从CEPS.csv数据里挑选若干变量建立回归方程，要求至少三个自变量
    + 如，学生的学业成绩受认知水平、家庭收入的影响
    + 考虑因变量和自变量间的实质关系，变量间关系应该是有意义
    + 选择自变量时，注意变量的类型，如果是分类变量，需要进行编码
+ 请报告回归方程的结果，需要包括：
    + 模型拟合指标
    + 模型的显著性检验结果
    + 变量的系数
    + 各系数的显著性检验结果
    + 对模型结果的解释
    

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
```

```python
#导入数据
sentinels = {'b11': [' '], 'b12': [' '], 'b13': [' '], 'c14': [' ']}
df = pd.read_csv('CEPS.csv',encoding='gb2312', na_values=sentinels)
sdf=df.sample(n=1000)
sdf
```

```python
data1= pd.DataFrame({
    'y': sdf.c14,
    'x1': sdf.b11,
    'x2': sdf.b12,
    'x3': sdf.b13,
})
#构建自变量
data1=data1.dropna(axis=0,how='any')#删除缺失值
model_x = ['x1','x2','x3']
X = data1.loc[ :,model_x].values
#构建因变量
y = data1['y'].values
#加入截距项
X_model = sm.add_constant(X)
print(X)
print(y)
print(X_model)
model = sm.OLS(y, X) 
#拟合ols回归
results = model.fit()
results.params

#输出结果
print(results.summary())
```

模型拟合指标：
因变量 y ：男生更擅长数学
自变量：
x1:有独立书桌
x2:家里书多
x3:有无电脑和网络
模型的显著性检验结果
P=0.000，有显著影响
变量的系数（表中coef对应值）
x1:0.6209
x2:0.1612
x3:0.0811
各系数的显著性检验结果（表中P对应值）
x1:p>0.05
x2:p>0.05
x3:p>0.05
由此看出四个变量对y均无显著影响，x1，x2，x3对男生是否更擅长数学的影响都不显著 三个自变量中与因变量之间都成正相关。
